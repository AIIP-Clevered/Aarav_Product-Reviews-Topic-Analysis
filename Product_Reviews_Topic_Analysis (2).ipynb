```python
def calculate_topic_cv(ldamodel,texts,dictionary,topic_range):
  cv_score =[]
  topic_num = []
  for i in range(2,topic_range):
    topic_num.append(i)
    Lda = gensim.models.ldamodel.LdaModel
    ldamodel = Lda(doc_term_matrix, num_topics=i, id2word=dictionary, passes=10,random_state=45)
    cv_score.append(CoherenceModel(model=ldamodel,texts=texts,\
                                   dictionary=dictionary , coherence='c_v').get_coherence())
    clear_output()
  return topic_num,cv_score
```

```python
import pandas as pd
import gensim
from gensim import corpora
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import ipywidgets as widgets
from IPython.display import display, clear_output
import warnings
import os

warnings.filterwarnings('ignore')

class KeywordReviewDashboard:
    def __init__(self):
        self.raw_data = None
        self.output = widgets.Output()

        # --- UI ELEMENTS ---
        self.file_input = widgets.Text(value='/content/K8 Reviews v0.2.xlsx', description='File Path:')
        self.keyword_input = widgets.Text(value='', placeholder='e.g. battery, camera', description='Search Topic:')
        self.topic_slider = widgets.IntSlider(value=5, min=2, max=10, description='Sub-Topics:')
        self.btn_run = widgets.Button(description="Analyze Search Topic", button_style='success')

        self.btn_run.on_click(self.on_run)

        display(widgets.VBox([
            widgets.HTML("<h2>üîç Topic-Specific Deep Dive</h2>"),
            widgets.HTML("<p>Type a word like <b>'battery'</b> to see what people specifically say about it.</p>"),
            self.file_input,
            self.keyword_input,
            self.topic_slider,
            self.btn_run,
            self.output
        ]))

    def on_run(self, b):
        path = self.file_input.value
        keyword = self.keyword_input.value.lower().strip()

        with self.output:
            clear_output()
            if not os.path.exists(path):
                print("‚ùå File not found!")
                return

            # 1. Load Data
            df = pd.read_excel(path) if path.endswith('.xlsx') else pd.read_csv(path)
            col = 'review' if 'review' in df.columns else df.columns[0]

            # 2. Filter by Keyword
            if keyword:
                filtered_df = df[df[col].astype(str).str.lower().str.contains(keyword)]
                print(f"üîé Found {len(filtered_df)} reviews mentioning '{keyword}'")
            else:
                filtered_df = df
                print(f"üìä Analyzing all {len(df)} reviews...")

            if len(filtered_df) < 10:
                print("‚ö†Ô∏è Too few reviews found for that word to create topics.")
                return

            # 3. Quick Clean & Process
            stop_words = set(stopwords.words('english'))
            lem = WordNetLemmatizer()
            processed = []
            for doc in filtered_df[col].fillna("").astype(str).str.lower():
                tokens = word_tokenize(doc)
                clean = [lem.lemmatize(w) for w in tokens if w not in stop_words and len(w)>3 and w.isalpha()]
                if clean: processed.append(clean)

            # 4. Generate LDA
            dictionary = corpora.Dictionary(processed)
            corpus = [dictionary.doc2bow(text) for text in processed]
            lda = gensim.models.ldamodel.LdaModel(corpus, num_topics=self.topic_slider.value, id2word=dictionary, passes=10)

            # 5. Display
            pyLDAvis.enable_notebook()
            display(pyLDAvis.display(gensimvis.prepare(lda, corpus, dictionary)))

dashboard = KeywordReviewDashboard()
```
